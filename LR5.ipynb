# Построить двунаправленную ассоциативную память, связывающую изображение из файла с его сжатым вариантом

```python
import numpy as np
from PIL import Image
```

```python
arrow = Image.open('arrow.bmp')
pressed_arrow = Image.open('pressed_arrow.bmp')
```

```python
gray = arrow.convert('L')
 
a1 = np.array(gray.getdata())
for i in range(len(a1)):
  if a1[i] > 0:
    a1[i] = 1
  else:
    a1[i] = -1
a1
```

```python
gray = pressed_arrow.convert('L')

a2 = np.array(gray.getdata())
for i in range(len(a2)):
  if a2[i] > 0:
    a2[i] = 1
  else:
    a2[i] = -1
a2
```

```python
w = np.zeros([len(a1), len(a2)])
for i in range(len(a1)):
  for j in range(len(a2)):
    w[i][j] = a1[i]*a2[j]
print('Матрица коэффициентов:')
w
```

```python
def Sign(a):
  for i in range(len(a)):
    if a[i] > 0:
      a[i] = 1
    else:
      a[i] = -1
  return a
```

```python
s = Sign(a1 @ w)
print(s)
count = 0
for i in range(len(a2)): 
  if s[i] == a2[i]:
    count += 1
    if count == len(a2):
      print('Полученный массив совпадает со сжатой картинкой')
```
