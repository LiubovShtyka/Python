2.Отобразите 3 графика - линейный, квадратичный и кубический на одном поле:

````python
x = np.linspace(-4, 4, 20)
y1 = x
# Квадратичная зависимость
y2 = [i**2 for i in x]
# Кубическая зависимость
y3 = [i**3 for i in x]
plt.plot(x, y1, x, y2, x, y3)  # построение графика
````
