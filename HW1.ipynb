Напишите программу на языке Python, которая будет рассчитывать частоту слов из текстового файла.

Вход: текстовый файл (https://drive.google.com/file/d/13AXtMaNQ8z2_80rq5kLsRX7HS5HpUwZ2/view?usp=sharing)

Выход: информацию о частоте встречаемости слов необходимо сохранить в виде словаря, затем записать в выходной файл по убыванию частоты.

```python
import re
```

```python
from google.colab import drive # Импорт файла из аккаунта

with open("/content/drive/My Drive/Нейросети/HW1/Garden.txt") as f: # Открываем файл
    for line in f:
      word_list = re.split('\s+', f.read().lower()) # Функция Re.Split () расщепляет данную строку в вхождении указанного шаблона.

print ('Words in text:', len(word_list))
```

```python
freq_dic = {}
punctuation = re.compile(r'[().…?!,":«»;—-]') 

for word in word_list:
    word = punctuation.sub("", word)
    try: 
        freq_dic[word] += 1
    except: 
        freq_dic[word] = 1 
print('Unique words:', len(freq_dic))
```

```python
# sort by key or word
freq_list = sorted(freq_dic.items()) #тут получается, что он сортирует именно по айтемам, что как бы понятно
# Поэтому сначала идут цифры, потом по буквам
freq_list.sort(reverse = True, key = lambda k: k[1])

f = open('Словарь.txt','w') 
for word, freq in freq_list:
    print(word, freq)
    f.write(word + " " + str(freq) + "\n")
f.close()
  
```
